### Resizing/Rescaling Images

```python
from skimage import color
from skimage.transform import rescale, resize, downscale_local_mean
```

```python
plt.imshow(data.astronaut())
plt.show()
```

```python
data.astronaut()
# What does each pixel (the three numbers) correspond to?
```

```python
image = color.rgb2gray(data.astronaut())
plt.imshow(image, cmap="gray")
plt.show()
```

```python
help(rescale)
```

```python
image_rescaled = rescale(image, 1.0 / 4.0, anti_aliasing=False)
plt.imshow(image_rescaled, cmap="gray")
plt.show()
```

```python
image_resized = resize(image, (image.shape[0] / 4, image.shape[1] / 4),
                       anti_aliasing=True)
plt.imshow(image_resized, cmap="gray")
plt.show()
```

```python
image_downscaled = downscale_local_mean(image, (4, 3))
plt.imshow(image_downscaled, cmap="gray")
plt.show()
```

### Masking Images

```python
import numpy
nrows, ncols = camera.shape
help(numpy.ogrid)
row, col = numpy.ogrid[:nrows, :ncols]
```

```python
cnt_row, cnt_col = nrows / 2, ncols / 2
outer_disk_mask = ((row - cnt_row)**2 + (col - cnt_col)**2 > (nrows / 2)**2)
outer_disk_mask
```

```python
camera[outer_disk_mask] = 0
```

```python
plt.imshow(camera)
plt.show()
```

```python
plt.imshow(camera, cmap="gray")
plt.show()
```

### Separating Color Channels in IHC

```python
from skimage.color import rgb2hed
from matplotlib.colors import LinearSegmentedColormap
```

```python
# Create an artificial color close to the original one
cmap_hema = LinearSegmentedColormap.from_list('mycmap', ['white', 'navy'])
cmap_dab = LinearSegmentedColormap.from_list('mycmap', ['white',
                                             'saddlebrown'])
cmap_eosin = LinearSegmentedColormap.from_list('mycmap', ['darkviolet',
                                               'white'])
```

```python
ihc_rgb = data.immunohistochemistry()
plt.imshow(ihc_rgb)
plt.show()
```

```python
rgb2hed?
```

```python
ihc_hed = rgb2hed(ihc_rgb)
plt.imshow(ihc_hed)
plt.show()
```

```python
plt.imshow(ihc_hed[:, :, 0], cmap=cmap_hema)
plt.show()
```

```python
plt.imshow(ihc_hed[:, :, 1], cmap=cmap_eosin)
plt.show()
```

```python
plt.imshow(ihc_hed[:, :, 2], cmap=cmap_dab)
plt.show()
```

```python
from skimage.exposure import rescale_intensity

# Rescale hematoxylin and DAB signals and give them a fluorescence look
h = rescale_intensity(ihc_hed[:, :, 0], out_range=(0, 1))
d = rescale_intensity(ihc_hed[:, :, 2], out_range=(0, 1))
zdh = numpy.dstack((np.zeros_like(h), d, h))

plt.imshow(zdh)
plt.show()
```

#### Edge-based Image Segmentation

```python
from skimage import filters
from scipy import ndimage as ndi
from skimage import feature

# Generate noisy image of a square
im = numpy.zeros((128, 128))
im[32:-32, 32:-32] = 1
plt.imshow(im, cmap="gray")
plt.show()
```

```python
im = ndi.rotate(im, 15, mode='constant')
im = ndi.gaussian_filter(im, 4)
im += 0.2 * np.random.random(im.shape)
plt.imshow(im, cmap="gray")
plt.show()
```

```python
# Compute the Canny filter for two values of sigma
edges1 = feature.canny(im)
edges2 = feature.canny(im, sigma=3)
```

```python
plt.imshow(edges1, cmap="gray")
plt.show()
```

```python
plt.imshow(edges2, cmap="gray")
plt.show()
```

```python
from skimage.feature import canny
edges = canny(coins/255.)
fill_coins = ndi.binary_fill_holes(edges)
```

```python
label_objects, nb_labels = ndi.label(fill_coins)
sizes = numpy.bincount(label_objects.ravel())
mask_sizes = sizes > 20
mask_sizes[0] = 0
coins_cleaned = mask_sizes[label_objects]
```

```python
plt.imshow(coins_cleaned, cmap="gray")
plt.show()
```

#### Region-based Image Segmentation

```python
markers = numpy.zeros_like(coins)
markers[coins < 30] = 1
markers[coins > 150] = 2
from skimage.filters import sobel
elevation_map = sobel(coins)
```

```python
plt.imshow(elevation_map)
plt.show()
```

```python
markers = numpy.zeros_like(coins)
markers[coins < 30] = 1
markers[coins > 150] = 2
markers
```

```python
plt.imshow(markers)
plt.show()
```

```python
from skimage.morphology import watershed
segmentation = watershed(elevation_map, markers)
```

```python
plt.imshow(segmentation)
plt.show()
```

```python
segmentation = ndi.binary_fill_holes(segmentation - 1)
labeled_coins, _ = ndi.label(segmentation)
```

```python
plt.imshow(labeled_coins)
plt.show()
```

```python
from skimage import measure
labels = measure.label(labeled_coins)
print(labels.max())
```

```python
print('coverage is %f' %(labeled_coins.mean()))
```
